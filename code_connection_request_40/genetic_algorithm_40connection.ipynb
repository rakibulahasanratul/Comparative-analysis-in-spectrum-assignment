```python
total_connection_40=[{'name': 'T1', 'p': 5, 'fixj': [67, 7, 41, 65], 'start': -1},
{'name': 'T2', 'p': 3, 'fixj': [50, 51, 23, 13, 54], 'start': -1},
{'name': 'T3', 'p': 4, 'fixj': [55, 37, 18, 10], 'start': -1},
{'name': 'T4', 'p': 2, 'fixj': [59, 61, 2, 16], 'start': -1},
{'name': 'T5', 'p': 3, 'fixj': [47, 68], 'start': -1},
{'name': 'T6', 'p': 5, 'fixj': [53, 18, 10, 12, 44], 'start': -1},
{'name': 'T7', 'p': 5, 'fixj': [27, 23, 64], 'start': -1},
{'name': 'T8', 'p': 5, 'fixj': [5, 55, 14, 23, 13, 54, 59, 61, 2, 16], 'start': -1},
{'name': 'T9', 'p': 4, 'fixj': [57, 44, 45, 61, 2, 16, 70], 'start': -1},
{'name': 'T10', 'p': 3, 'fixj': [45, 61], 'start': -1},
{'name': 'T11', 'p': 4, 'fixj': [36, 14, 23, 13, 54, 56], 'start': -1},
{'name': 'T12', 'p': 1, 'fixj': [48, 26, 19, 42, 21, 58, 28, 3, 9, 43, 24], 'start': -1},
{'name': 'T13', 'p': 3, 'fixj': [69, 6, 14, 23, 13, 54, 56], 'start': -1},
{'name': 'T14', 'p': 4, 'fixj': [38, 44, 45, 61], 'start': -1},
{'name': 'T15', 'p': 2, 'fixj': [28, 3, 9, 43, 15], 'start': -1},
{'name': 'T16', 'p': 2, 'fixj': [36, 14, 23, 13, 54, 59, 61, 2, 16, 46], 'start': -1},
{'name': 'T17', 'p': 2, 'fixj': [48, 26, 19, 42, 21, 58, 28, 3, 9, 43, 66, 17], 'start': -1},
{'name': 'T18', 'p': 5, 'fixj': [54, 59, 29], 'start': -1},
{'name': 'T19', 'p': 4, 'fixj': [25, 30, 6, 37], 'start': -1},
{'name': 'T20', 'p': 3, 'fixj': [31, 28, 3, 9, 43, 15], 'start': -1},
{'name': 'T21', 'p': 2, 'fixj': [53, 39, 14, 23, 13], 'start': -1},
{'name': 'T22', 'p': 3, 'fixj': [3, 9, 43, 37, 18, 73], 'start': -1},
{'name': 'T23', 'p': 5, 'fixj': [35, 67, 7, 41, 47, 60], 'start': -1},
{'name': 'T24', 'p': 1, 'fixj': [43], 'start': -1},
{'name': 'T25', 'p': 4, 'fixj': [38, 44, 45, 58, 56], 'start': -1},
{'name': 'T26', 'p': 3, 'fixj': [28, 3, 9, 43, 37], 'start': -1},
{'name': 'T27', 'p': 1, 'fixj': [71, 18, 10], 'start': -1},
{'name': 'T28', 'p': 5, 'fixj': [20, 62, 28, 3, 9, 43, 24], 'start': -1},
{'name': 'T29', 'p': 2, 'fixj': [4, 32, 40, 22, 72], 'start': -1},
{'name': 'T30', 'p': 2, 'fixj': [6, 37], 'start': -1},
{'name': 'T31', 'p': 4, 'fixj': [51, 23, 13, 54, 59, 61, 2, 16, 46, 63], 'start': -1},
{'name': 'T32', 'p': 5, 'fixj': [9, 43, 66], 'start': -1},
{'name': 'T33', 'p': 4, 'fixj': [71, 39, 24], 'start': -1},
{'name': 'T34', 'p': 3, 'fixj': [25, 27, 23, 64], 'start': -1},
{'name': 'T35', 'p': 3, 'fixj': [33, 19], 'start': -1},
{'name': 'T36', 'p': 3, 'fixj': [42, 49, 1], 'start': -1},
{'name': 'T37', 'p': 3, 'fixj': [25, 27, 23, 13, 54, 8, 34], 'start': -1},
{'name': 'T38', 'p': 1, 'fixj': [48, 26, 19, 42, 49], 'start': -1},
{'name': 'T39', 'p': 3, 'fixj': [11, 32, 39, 66, 17], 'start': -1},
{'name': 'T40', 'p': 3, 'fixj': [52, 23], 'start': -1}];


import random
import time
import itertools
from matplotlib import pyplot as plt
import numpy as np


#Initialization of Class Task containing class name, processing time "p" and list of fixed processor "fixj"
class Task:
    def __init__(self, name, p, fixj):
        self.name = name
        self.processing_time = p
        self.fixed_processors = fixj

#Initialization of Class Dataset to get task details from the input task details

class Dataset:
    def __init__(self, tasks):
        self.tasks = [Task(task['name'], task['p'], task['fixj']) for task in tasks] # Iterate over tasks input and convert it to a Task objet Class Task.
        self.n_tasks = len(tasks) # calculate the length of the tasks list.
        self.processors = list(set(itertools.chain.from_iterable(task['fixj'] for task in tasks))) #for given processor associated to the task "itertools.chain.from_iterable" remove the duplicate processor and return to a list
        self.n_processors = len(self.processors) #calculating the number of unique processors
        self.time_matrix = self._create_time_matrix()

# Helper function "_create_time_matrix" to derive the time_matrix in the class dataset.
    def _create_time_matrix(self):
        time_matrix = [] #initialization
        for task in self.tasks:
            row = [float('inf')] * self.n_processors #for each task of all processors is updated to float('inf')
            for processor_id in task.fixed_processors: #Identification of task respective fixed processor
                processor_idx = self.processors.index(processor_id) #Index for identified fixed processor
                row[processor_idx] = task.processing_time #updating the fixed processor with processing time.
            time_matrix.append(row)
        return time_matrix

#Genetic Algorithms..............

class GeneticTaskScheduler:
    def __init__(self, population_size, crossover_rate, mutation_rate, max_repeats):
        self.population_size = population_size #the number of individual population
        self.crossover_rate = crossover_rate # probability of crossover between individuals
        self.mutation_rate = mutation_rate # probability of mutation in individuals
        self.max_repeats = max_repeats #the max number of generation without improvement before stopping.
        self.statistics = []

    #defining the initial population set of individuals
    def initialize_population(self, dataset):
      population = []
      task_indices = list(range(dataset.n_tasks))  # List of task indices
      for _ in range(self.population_size):
        random.shuffle(task_indices)  # Shuffle the task order
        population.append(task_indices[:])  # Add a copy of the shuffled order
      return population


    # Calculating fitness of each schedule or individual. Objective is to minimize the processing time.

    def fitness(self, individual, dataset):
    # Initialize processor availability times
      processor_times = {processor: 0 for processor in dataset.processors}
    # Go through each task in the individual schedule
      for task_id, _ in enumerate(individual):
        task = dataset.tasks[task_id]
        max_time = 0
        # Find the time when all processors for this task are available
        for processor in task.fixed_processors:
            max_time = max(max_time, processor_times[processor])
        # Schedule the task at max_time and update processor availability times
        for processor in task.fixed_processors:
            processor_times[processor] = max_time + task.processing_time
    # The fitness is the time when the last processor becomes available
      return max(processor_times.values())

    #the selection process is a min process to find the individual has the lowest fitness score
    def selection(self, population, dataset):
        return min(population, key=lambda individual: self.fitness(individual, dataset))

    def crossover(self, parent1, parent2, dataset):
      point = random.randint(1, dataset.n_tasks - 2)
      child1_part1 = parent1[:point]
      child1_part2 = [task for task in parent2 if task not in child1_part1]
      child2_part1 = parent2[:point]
      child2_part2 = [task for task in parent1 if task not in child2_part1]
      return child1_part1 + child1_part2, child2_part1 + child2_part2


    def mutation(self, individual, dataset):
      mutated_individual = individual[:]
      if len(mutated_individual) > 1:  # Ensure there are at least two tasks to swap
        idx1, idx2 = random.sample(range(len(mutated_individual)), 2)
        # Swap two tasks
        mutated_individual[idx1], mutated_individual[idx2] = mutated_individual[idx2], mutated_individual[idx1]
      return mutated_individual


    def optimize(self, dataset):
        population = self.initialize_population(dataset)
        best_individual = self.selection(population, dataset)
        best_fitness = self.fitness(best_individual, dataset)
        repeats = 0

        while repeats < self.max_repeats:
            new_population = [best_individual]
            while len(new_population) < self.population_size:
                parent1, parent2 = random.sample(population, 2)
                if random.random() < self.crossover_rate:
                    child1, child2 = self.crossover(parent1, parent2, dataset)
                else:
                    child1, child2 = parent1, parent2
                if random.random() < self.mutation_rate:
                    child1 = self.mutation(child1, dataset)
                if random.random() < self.mutation_rate:
                    child2 = self.mutation(child2, dataset)
                new_population.extend([child1, child2])
            population = new_population[:self.population_size]
            current_best = self.selection(population, dataset)
            current_fitness = self.fitness(current_best, dataset)
            if current_fitness < best_fitness:
                best_fitness = current_fitness
                best_individual = current_best
                repeats = 0
            else:
                repeats += 1
            self.statistics.append(best_fitness)

        return best_individual

    def calculate_start_times(self, best_solution, dataset):
    # Initialize the next available time for each processor
      next_available_time = {processor: 0 for processor in dataset.processors}

    # Initialize the start times for each task
      start_times = {}

      for task_id in best_solution:
        task = dataset.tasks[task_id]
        start_time = 0

        # Find the next available time slot where all processors for this task are available
        for processor in task.fixed_processors:
            start_time = max(start_time, next_available_time[processor])

        # Schedule the task at this start time
        end_time = start_time + task.processing_time
        start_times[task.name] = start_time

        # Update the next available time for all processors required by this task
        for processor in task.fixed_processors:
            next_available_time[processor] = end_time

      return start_times

    def schedule(self, dataset):
      start_time = time.time()
      best_solution = self.optimize(dataset)
      elapsed_time = time.time() - start_time

    # Calculate start times for each task
      start_times = self.calculate_start_times(best_solution, dataset)

    # Print tabular data
      self.print_tabular_data(start_times)

    # Gantt chart visualization
      #self.plot_gantt_chart(start_times, dataset)

     # Gantt chart visualization
      self.plot_gantt_chart2(start_times, dataset)

      print(f"Best fitness: {self.fitness(best_solution, dataset)}")

    def print_tabular_data(self, start_times):
      print("\nTask Scheduling Table:")
      print("Task\tScheduled Start Time")
      for task_name in sorted(start_times.keys()):
        print(f"{task_name}\t{start_times[task_name]}")

    def plot_gantt_chart(self, start_times, dataset):
      fig, ax = plt.subplots(figsize=(15, 7))

      processors_colors = plt.cm.get_cmap("tab10", dataset.n_processors)
      processor_colors_dict = {processor: processors_colors(i) for i, processor in enumerate(dataset.processors)}

      yticks = []
      yticklabels = []

      for task_idx, (task_name, start_time) in enumerate(start_times.items()):
        task = next(task for task in dataset.tasks if task.name == task_name)
        processor_colors = [processor_colors_dict[p] for p in task.fixed_processors]

        ax.barh(task_idx, task.processing_time, left=start_time, color=processor_colors, edgecolor='black')
        yticks.append(task_idx)
        yticklabels.append(f"{task_name} (Proc: {', '.join(map(str, task.fixed_processors))})")

      ax.set_xlabel('Wavelength Assignment in each shortest path')
      ax.set_ylabel('Shortest Path with edge details (Total:40)')
      ax.set_title('Gantt Chart - Task Scheduling')
      ax.set_yticks(yticks)
      ax.set_yticklabels(yticklabels)

    # Set custom x-axis ticks and labels
      end_times = {task_name: start_time + dataset.tasks[next(i for i, task in enumerate(dataset.tasks) if task.name == task_name)].processing_time for task_name, start_time in start_times.items()}
      ticks = list(range(0, max(end_times.values()) + 1, 1))  # Adjust the step value as needed
      ax.set_xticks(ticks)
      ax.set_xticklabels(ticks)

      plt.show()

    def plot_gantt_chart2(self, start_times, dataset):
      fig_width = max(7, 0.5 * len(dataset.processors))
      fig, ax = plt.subplots(figsize=(fig_width, 20))

      # Assign a color to each task for visual distinction
      task_colors = plt.cm.get_cmap("tab10", len(dataset.tasks))
      task_color_dict = {task.name: task_colors(i) for i, task in enumerate(dataset.tasks)}

      # Create a dictionary to map processors to X-axis positions
      processor_x_pos = {processor: i for i, processor in enumerate(dataset.processors)}
      processor_labels = [f'P{processor}' for processor in dataset.processors]

      # Plot each task on the Gantt chart
      for task_name, start_time in start_times.items():
          task = next(t for t in dataset.tasks if t.name == task_name)
          # Each task is given a vertical bar in the Gantt chart
          for processor in task.fixed_processors:
              x_pos = processor_x_pos[processor]
              ax.bar(x_pos - 0.4, task.processing_time, bottom=start_time, width=0.8,
                    color=task_color_dict[task_name], edgecolor='black')

      # Set the labels and title of the Gantt chart
      ax.set_ylabel('Wavelength Assignment in each shortest path', fontsize=20)
      ax.set_xlabel('Link in the shortest Path', fontsize=20)
      ax.set_title('Processor schedule for 40 connection request', fontsize=20)
      ax.set_xticks(list(processor_x_pos.values()))
      ax.set_xticklabels(processor_labels)

      # Determine the end time for the y-axis limit
      max_end_time = max(start_time + task.processing_time for task_name, start_time in start_times.items())
      ax.set_yticks(range(0, max_end_time + 1, 1))

      # Show the Gantt chart
      plt.tight_layout()
      plt.show()




# Define the input tasks
#tasks_input = updated_tasks
tasks_input=total_connection_40

# Create the dataset from the input
dataset = Dataset(tasks_input)

# Initialize the genetic algorithm
ga = GeneticTaskScheduler(population_size=10, crossover_rate=0.7, mutation_rate=0.2, max_repeats=10000)

# Run the scheduling
ga.schedule(dataset)
```
